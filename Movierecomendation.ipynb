# **Movie Recommendation System**

**Recomendation system** is a system that seeks to predict or filter preferences acording to the user's choices recomender system are utilised in variety of area including movies , music , news , books , research articles search queries, social tags products in general Recomender system produce a list of recommendation in any of the two ways -

**Collaborative filtering:** collaborative filtering apporaches built a model from the user's past behavior (i.e. items purchased or searched by the user )as well as similar decision made by other users. This model is then used to predict items (or ratings for items ) that users may have an intrests in.

**Content-based filtering:** Content-based filtering approaches uses a series of discrete characteristics of an item in order to recomended additional items with similar properties. Content-based filtering methods are totattly based on a discription of the item an a profile of the user's preferences it recomended item based on user's past preferences . let's develop a basic recomendation system using python and pandas .

# **Get Feature Selection**

```python
df_features = df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast','Movie_Director']].fillna('')
```

Selected Five existing features to recommend movies. it may vary from one project to another like one can add vote counts,budget,language, etc

```python
x = df_features['Movie_Genre']+' '+df_features['Movie_Keywords']+' '+df_features['Movie_Tagline']+' '+df_features['Movie_Cast']+' '+df_features['Movie_Director']
```

# **Get Feature Text Conversion to Tokens**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x=tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

# **Get Similarity Score Using Cosine Similarity**

cosine_similarity computes the l2-normalised dot product of vectors . Euclidean (L2) normalisation projects the vectors onto the sphere ant their dot product is thene the cosine of the angle between the point denoted by the vectors.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score= cosine_similarity(x)
```

```python
Similarity_score
```

```python
Similarity_score.shape
```

# **Get Movie Name as Input From User and Validate For Closest Spelling**

```python
Favourite_Movie_Name = input('Enter your favourite movie name :')
```

```python
All_Movies_Title_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_recommendation = difflib.get_close_matches(Favourite_Movie_Name,All_Movies_Title_List)
print(Movie_recommendation)
```

```python
Close_Match = Movie_recommendation[0]
print(Close_Match)
```

```python
Index_of_Close_Match_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
print(Index_of_Close_Match_Movie)
```

```python
#getting a list of similar movie
recomendation_score = list(enumerate(Similarity_score[Index_of_Close_Match_Movie]))
print(recomendation_score)
```

```python
len(recomendation_score)
```

# **Get All Movie Sort Based On Recomendation Score wrt Favourite Movie**

```python
#sorting the movie based on theri similarity score
sorted_similar_movies = sorted(recomendation_score,key = lambda x:x[1],reverse = True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index
print('Top 30 Movies Suggested For You : \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['Movie_Title'].values[0]
  if (i<31):
    print(i,'.',title_from_index)
    i+=1
```

# **Top 10 Movie Recomendation System **

```python
Movie_name = input('Enter your favourite movie name :')
list_of_all_titles = df['Movie_Title'].tolist()
Find_Close_Match = difflib.get_close_matches(Movie_name,list_of_all_titles)
Close_Match = Find_Close_Match[0]
Index_of_Movie = df[df.Movie_Title == Close_Match]['Movie_ID'].values[0]
Recommendation_Score = list(enumerate(Similarity_score[Index_of_Movie]))
sorted_similar_movies = sorted(Recommendation_Score,key = lambda x:x[1],reverse = True)
print('Top 10 Movies Suggested For You : \n')
i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.Movie_ID==index]['Movie_Title'].values
  if (i<11):
    print(i,'.',title_from_index)
    i+=1
```
